```python
earning = pd.read_csv('C:/Users/subao/Desktop/zheshang/earnings.csv')
accrual = pd.read_csv('C:/Users/subao/Desktop/zheshang/accrual.csv')
cash = pd.read_csv('C:/Users/subao/Desktop/zheshang/cash.csv')
asset = pd.read_csv('C:/Users/subao/Desktop/zheshang/asset.csv')

earning = earning.drop(axis = 1, columns=['Unnamed: 0'])
accrual = accrual.drop(axis = 1, columns=['Unnamed: 0'])
cash = cash.drop(axis = 1, columns=['Unnamed: 0'])
asset = asset.drop(axis = 1, columns=['Unnamed: 0'])
code = earning['code']
```

```python
earning_r = DataFrame(code, columns = ['code'])
accrual_r = DataFrame(code, columns = ['code'])
cash_r = DataFrame(code, columns = ['code'])

for n in range(1,12):
    earning_r[earning.columns[n]] = earning.iloc[:,n] / asset.iloc[:,n]
    accrual_r[accrual.columns[n]] = accrual.iloc[:,n] / asset.iloc[:,n]
#    cash_r[cash.columns[n]] = cash.iloc[:,n] / asset.iloc[:,n]
    cash_r[cash.columns[n]] = earning_r[earning.columns[n]] - accrual_r[accrual.columns[n]]
```
